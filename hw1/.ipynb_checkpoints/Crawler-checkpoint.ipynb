## Web Crawler

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd

start_day = 1451577600       # Dec 31, 2015
end_day = int(time.time())   # Now()
per_day = 86400
month_trans_dict = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}

my_options = Options()
my_options.add_argument("--headless")
#my_options.add_argument("download.default_directory=.")
driver = webdriver.Chrome('./chromedriver.exe', chrome_options=my_options)
#driver = webdriver.Chrome('./chromedriver.exe')

def get_statistics(etf_name, get_time):
    '''
    In this function, we crawl the data of each etf by the name.
    Due to many complicating reasons, what I do in follow is:
        crawl from 12/31/2017 back to 12/31/2015,
        each iteration I only get 90 days, and use last_day_of_stock_market_dict to filter the valid days.
        records the needed NAV, and discards everything else.
    '''
    print('#### start handle %s ####'%(etf_name))
    statistic = []
    all_time_list = []
    period2 = end_day
    period1 = end_day - per_day * 90
    while True:
        break_flag = False
        if period1 <= start_day:
            period1 = start_day
            break_flag = True
        url = 'https://finance.yahoo.com/quote/%s/history?period1=%d&period2=%d&interval=1d&filter=history&frequency=1d'%(etf_name, period1, period2)
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, features='lxml')
        tr_list = soup.find_all('tr', {"class": "BdT Bdc($c-fuji-grey-c) Ta(end) Fz(s) Whs(nw)"})
        for tr in tr_list:
            td_list = tr.find_all('td', {'class': 'Py(10px) Pstart(10px)'})
            if len(td_list) < 1:
                continue
            if get_time:
                td_time = tr.find('td', {'class': 'Py(10px) Ta(start) Pend(10px)'})
                time_list = re.split(r'[,\s]', td_time.get_text())
                time_format = '%04d-%02d-%02d'%(int(time_list[3]), month_trans_dict[time_list[0]], int(time_list[1]))
                all_time_list.append(time_format)
                #print(time_format)
            statistic.append(float(td_list[4].get_text()))
            #print([td.get_text() for td in td_list])
        if break_flag:
            break
        period2 = period1
        period1 = period2 - per_day * 90
        #print('-----')
    print(len(statistic), len(all_time_list))
    return statistic, all_time_list

def get_all_statistics(etf_name_list):
    statistics = []
    extra_etf = ['FENY', 'TPYP']
    get_time = True
    dataFrame_time = []
    dataFrame_column = []
    #for etf_name in etf_name_list:
    for etf_name in ['XLE']:
        statistic, dataFrame_time = get_statistics(etf_name, get_time)
        if etf_name in extra_etf:
            # extra handling
            continue
        dataFrame_column.append(etf_name)
        statistics.append(statistic)
        if get_time:
            get_time = False
    print(statistics)
    print(dataFrame_column)
    print(dataFrame_time)
    statistics_dataFrame = pd.DataFrame(statistics, columns = dataFrame_column, index = dataFrame_time)
    return statistics
    
```
