```python
import pandas as pd
import numpy as np
A = pd.read_csv('rank_with_functionA.csv')
B_C_week = pd.read_csv('rank_with_functionB&C_week.csv')
B_C_month = pd.read_csv('rank_with_functionB&C_month.csv')
```

```python
Rank_week = pd.DataFrame()
Rank_month = pd.DataFrame()
Rank_week["ASSR_week_b2_rank"] = A["ASSR_week_b2_rank"]
Rank_week["ASSR_week_b1_rank"] = A["ASSR_week_b1_rank"]
Rank_week["Sharpe_Omega_week_rank"] = B_C_week["Sharpe_Omega_week_rank"]
Rank_week["Riskiness_week_uniform_rank"] = B_C_week["Riskiness_week_uniform_rank"]
Rank_week["Riskiness_week_fitpdf_rank"] = B_C_week["Riskiness_week_fitpdf_rank"]

Rank_month["ASSR_month_b2_rank"] = A["ASSR_month_b2_rank"]
Rank_month["ASSR_month_b1_rank"] = A["ASSR_month_b1_rank"]
Rank_month["Sharpe_Omega_month_rank"] = B_C_month["Sharpe_Omega_month_rank"]
Rank_month["Riskiness_month_uniform_rank"] = B_C_month["Riskiness_month_uniform_rank"]
Rank_month["Riskiness_month_fitpdf_rank"] = B_C_month["Riskiness_month_fitpdf_rank"]

Rank_month.to_csv("Rank_month.csv")
Rank_week.to_csv("Rank_week.csv")
```
