```python
import datetime
week_date = []
month_date = []
week_start = None
month_start = None

def get_weekday(date_string):
    date_list = list(map(int, date_string.split('-')))
    date_weekday = datetime.date(date_list[0], date_list[1], date_list[2]).weekday()
    return date_weekday

def get_month(date_string):
    date_list = list(map(int, date_string.split('-')))
    return date_list[1]
  
# main
for idx, date_string in enumerate(df_all['Date']):
    if idx == len(df_all['Date']) - 1:
        continue
    if get_weekday(df_all['Date'][idx]) > get_weekday(df_all['Date'][idx + 1]):
        week_date.append([df_all['Date'][idx], idx])
    if get_month(df_all['Date'][idx]) != get_month(df_all['Date'][idx + 1]):
        month_date.append([df_all['Date'][idx], idx])

print(week_date)
print(len(week_date))
print(month_date)
print(len(month_date))

```

```python
week_data_all = {}
month_data_alll = {}
for etf in df_all.columns:
    if etf == 'Date':
        continue
    week_data = []
    month_data = []
    for idx, (_, week_end_idx) in enumerate(week_date):
        if idx == 0:
            continue
        this_week = float(df_all[etf][week_end_idx])
        last_week = float(df_all[etf][week_date[idx-1][1]])
        week_data.append( ( this_week - last_week ) / this_week )
    for idx, (_, month_end_idx) in enumerate(month_date):
        if idx == 0:
            continue
        this_month = float(df_all[etf][month_end_idx])
        last_month = float(df_all[etf][month_date[idx-1][1]])
        month_data.append( ( this_month - last_month ) / this_month )
        
    week_data_all[etf] = week_data
    month_data_alll[etf] = month_data

df_week_data = pd.DataFrame(week_data_all)
df_month_data = pd.DataFrame(month_data_alll)

df_week_data.to_csv('df_week_data.csv')
df_month_data.to_csv('df_month_data.csv')
```
